```python
 import openpyxl
from IPython.display import FileLink
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# قراءة الداتا$$$$$$$AHMED GHORAIB$$$$$$$$$$ بـ openpyxl
wb = openpyxl.load_workbook('data2-2.xlsx')   
ws = wb.active
header = [cell.value for cell in ws[1]]

print("🔍 تحليل حسب البراند فقط...")

#$$$$$$$AHMED GHORAIB$$$$$$$$$ تحديد الأعمدة
col_indices = {col: i+1 for i, col in enumerate(header)}

competitors = ['Maxab', 'Waffarlk', 'Speed',
               'Al Yasmeen', 'Al Amir', 'Al Tarek', 'Al Muassasa']

your_price_idx = col_indices['tawfeerPrice']
manufacturer_idx = col_indices['manufacturerName']

def to_float(v):
    try:
        return float(v) if v not in (None, "") else None
    except:
        return None

# قوائم الإحصائيات $$$$$$$$$Ahmed Ghoraib(بالبراند فقط)
first_cheapest = []
second_cheapest = []
third_cheapest = []
most_expensive = []
only_me = []

for row in ws.iter_rows(min_row=2, values_only=True):
    my_price = to_float(row[your_price_idx-1])
    if not my_price or my_price == 0:
        continue

    comp_prices = [to_float(row[col_indices[c]-1]) for c in competitors if c in col_indices]
    comp_prices = [p for p in comp_prices if p and p > 0]

    # عندي أنا بس $$$$$$$$Ahmed Ghoraib$$$$$$$$$$$$$(كل المنافسين 0 أو فاضي)
    if not comp_prices:
        only_me.append({'brand': row[manufacturer_idx-1]})
        continue

    #$$$$AHMED GHORAIB$$$$$ أرخص واحد
    if my_price < min(comp_prices):
        first_cheapest.append({'brand': row[manufacturer_idx-1]})

    # $$$$$AHMED GHORAIB$$$$$تاني أرخص
    elif len(comp_prices) >= 2:
        sorted_comp = sorted(comp_prices)
        if my_price > sorted_comp[0] and my_price <= sorted_comp[1]:
            second_cheapest.append({'brand': row[manufacturer_idx-1]})

    # $$$$$$$AHMED GHORAIB$$$$$تالت أرخص
    elif len(comp_prices) >= 3:
        sorted_comp = sorted(comp_prices)
        if my_price > sorted_comp[1] and my_price <= sorted_comp[2]:
            third_cheapest.append({'brand': row[manufacturer_idx-1]})

    # $$$$$$AHMED GHORAIB$$$$أغلى واحد
    elif my_price > max(comp_prices):
        most_expensive.append({'brand': row[manufacturer_idx-1]})

# =============== تجميع الإحصائيات بالبراند فقط =====$$$$$Ahmed Ghoraib==========
def get_stats_by_brand(data_list, label_name):
    stats = {}
    for item in data_list:
        brand = item['brand']
        stats[(brand, label_name)] = stats.get((brand, label_name), 0) + 1

    result = []
    for (brand, label), count in stats.items():
        result.append({
            'البراند': brand,
            'عدد_المنتجات': count,
            'الفئة': label
        })
    return result

first_stats  = get_stats_by_brand(first_cheapest,  ' أرخص واحد')
second_stats = get_stats_by_brand(second_cheapest, ' تاني أرخص')
third_stats  = get_stats_by_brand(third_cheapest,  ' تالت أرخص')
most_stats   = get_stats_by_brand(most_expensive,  ' أغلى واحد')
only_stats   = get_stats_by_brand(only_me,         ' عندي بس')

all_stats = first_stats + second_stats + third_stats + most_stats + only_stats

print("\n📊 الإحصائيات (براند فقط):")
for stat in all_stats[:10]:
    print(f"{stat['الفئة']} - {stat['البراند']}: {stat['عدد_المنتجات']} منتج")

# =============== رسمتين$$$$$$$$Ahmed Ghoraib$$$$$$$$ Plotly جنب بعض ===============
colors_map = {
    '🥇 أرخص واحد': '#00ff88',
    '🥈 تاني أرخص': '#ffaa00',
    '🥉 تالت أرخص': '#ff6600',
    '🔴 أغلى واحد': '#ff4444',
    '🔵 عندي بس':  '#4444ff'
}

labels = ['🥇 أرخص واحد', '🥈 تاني أرخص', '🥉 تالت أرخص', '🔴 أغلى واحد', '🔵 عندي بس']
counts = [
    len(first_cheapest),
    len(second_cheapest),
    len(third_cheapest),
    len(most_expensive),
    len(only_me)
]

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('توزيع الفئات (Pie)', 'عدد المنتجات لكل فئة (Bar)'),
    specs=[[{"type": "pie"}, {"type": "bar"}]]
)

# $$$$$$$$$$AHMED GHORAIB$$$$$$$Pie
fig.add_trace(
    go.Pie(
        labels=labels,
        values=counts,
        marker=dict(colors=[colors_map[l] for l in labels]),
        textinfo='label+percent+value'
    ),
    row=1, col=1
)

#$$$$$$$AHMED GHORAIB$$$$$$ Bar
fig.add_trace(
    go.Bar(
        x=labels,
        y=counts,
        marker_color=[colors_map[l] for l in labels],
        text=counts,
        textposition='auto'
    ),
    row=1, col=2
)

fig.update_layout(
    height=600,
    title="تحليل الفئات السعرية حسب البراند",
    showlegend=False
)
fig.show()

fig.write_html('pricing_brand_analysis.html')
fig.write_image('pricing_brand_analysis.png')

# =============== حفظ $$$$$$$$$$$$$$$Ahmed Ghoraib$$$$$$$في Excel ===============
output_wb = openpyxl.Workbook()
ws1 = output_wb.active
ws1.title = "🥇 أرخص واحد"

def write_stats_to_sheet(ws, stats_list):
    ws['A1'] = "البراند"
    ws['B1'] = "عدد المنتجات"
    row_num = 2
    for stat in stats_list:
        ws[f'A{row_num}'] = stat['البراند']
        ws[f'B{row_num}'] = stat['عدد_المنتجات']
        row_num += 1 

write_stats_to_sheet(ws1, first_stats)
ws2 = output_wb.create_sheet("🥈 تاني أرخص")
write_stats_to_sheet(ws2, second_stats)
ws3 = output_wb.create_sheet("🥉 تالت أرخص")
write_stats_to_sheet(ws3, third_stats)
ws4 = output_wb.create_sheet("🔴 أغلى واحد")
write_stats_to_sheet(ws4, most_stats)
ws5 = output_wb.create_sheet("🔵 عندي بس")
write_stats_to_sheet(ws5, only_stats)

output_wb.save('pricing_by_brand_only.xlsx')
print("\n✅ تم الحفظ: pricing_by_brand_only.xlsx")
FileLink('pricing_by_brand_only.xlsx')
#كود مش عارف يقرا البيانات صح وتم العمل على تعديله والتعديل شامل كود اللئيمة كارلوس بتاع العبدلله يعنى ماتحاولش تسرقه علشان هيبهدل الداتا بتاعك وهيوديك فى خسارة ماتسرقش تعبى بس ماتهونش علبا اغرقك بردو فقلت انصحك
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$AhmedGhoraib$$$$$$$$$$$$$$$$$$$$$$ البيانات =========
df = pd.read_excel('data2-2.xlsx')

competitors = ['Maxab', 'Waffarlk', 'Speed',
               'Al Yasmeen', 'Al Amir', 'Al Tarek', 'Al Muassasa']
your_col = 'tawfeerPrice'

# $$$$$$$$$$$AHMED Ghoraib$$$$$$$$$$$$تحويل الأسعار لأرقام
for c in competitors + [your_col]:
    df[c] = pd.to_numeric(df[c], errors='coerce')

# ========= 2)$$$$$$$$$$$$$$$$$$$Ahmed Ghoraib$$$$$$$$$$$$$دالة التصنيف (لحد خامس + أغلى + وسط + عندك بس) =========
def classify_price_row(row):
    """
    ترجع ترتيب سعرك وسط (إنت + المنافسين > 0):
    1,2,3,4,5,'max','middle','only_you' أو None
    """
    my_price = row[your_col]
    comp_prices = [row[c] for c in competitors if pd.notna(row[c]) and row[c] > 0]

    # $$$$$$$$$$AHMED GHORAIB$$$$$$$$$$$$ولا عندك ولا عندهم سعر
    if (pd.isna(my_price) or my_price <= 0) and len(comp_prices) == 0:
        return None

    # $$$$$$$$$$$$$$$$$$$$AHMED GHORAIB$$$$$$$$$$$$عندى انا بس
    if (pd.notna(my_price) and my_price > 0) and len(comp_prices) == 0:
        return 'only_you'

    # عند المنافسين سعر وسعرك مش موجود$$$$$$$$$$$$$$$$$$$$$$$$$Ahmed Ghoraib$$$$$$$$$$ / 0
    if (pd.isna(my_price) or my_price <= 0) and len(comp_prices) > 0:
        return None

    #$$$$$$$$$$$$$$$$$$$$Ahmed Ghoraib$$$$$$$$$$$$$$$ عندى سعر والمنافسين عندهم سعر
    all_prices = comp_prices + [my_price]
    all_sorted = sorted(all_prices)
    rank = all_sorted.index(my_price) + 1
    max_price = all_sorted[-1]

    if rank == 1:
        return 1
    elif rank == 2:
        return 2
    elif rank == 3:
        return 3
    elif rank == 4:
        return 4
    elif rank == 5:
        return 5
    elif my_price == max_price:
        return 'max'
    else:
        return 'middle'

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$Ahmed Ghoraib$$$$$$$$$$$$$$$$$ تطبيق التصنيف
df['rank_you'] = df.apply(classify_price_row, axis=1)

# ========= 3)$$$$$$$$$$$$ملخص الأعداد والنِّسَب$$$$Ahmed Ghoraib$$$$$$$$$$$$$$$ =========
total_rows = len(df)
classified = df['rank_you'].notna().sum()
unclassified = total_rows - classified
 
print("📌 إجمالي المنتجات في الشيت:", total_rows)
print("📌 المنتجات اللي اتصنّفت:", classified)
print("📌 منتجات خارج كل الفئات:", unclassified)

order = [1, 2, 3, 4, 5, 'max', 'middle', 'only_you']
labels_map = {
    1: '🥇 أرخص',
    2: '🥈 تاني',
    3: '🥉 تالت',
    4: '4️⃣ رابع',
    5: '5️⃣ خامس',
    'max': '🔴 أغلى',
    'middle': '⚪ وسط',
    'only_you': '🔵 عندي بس'
}
counts = (
    df['rank_you']
      .value_counts()
      .reindex(order)
      .fillna(0)
      .astype(int)
)

labels = [labels_map[k] for k in order]
vals = counts.values.tolist()

print("\n📊 الأعداد:")
for lab, v in zip(labels, vals):
    print(f"{lab}: {v}")
    #كود اللئيمة$$$$$$$$$$$$$Ahmed Ghoraib$$$$$$$$$$$$$$$$$$$$$$$$$$$$ بتاعة بطاطا
output_wb = openpyxl.Workbook()
ws1 = output_wb.active
ws1.title = "🥇 أرخص واحد"

def write_stats_to_sheet(ws, stats_list):
    ws['A1'] = "البراند"
    ws['B1'] = "عدد المنتجات"
    row_num = 2
    for stat in stats_list:
        ws[f'A{row_num}'] = stat['البراند']
        ws[f'B{row_num}'] = stat['عدد_المنتجات']
        row_num += 1

write_stats_to_sheet(ws1, first_stats)         #$$$$$$$$$$$$$$$$كود اللئيمة بتاعة كارلوس$$$$$$$Ahmed Ghoraib$$$$$$$$ 
output_wb.save('pricing_by_brand_only.xlsx')
print("\n✅ تم  الحفظ: pricing_by_brand_only.xlsx")
FileLink('pricing_by_brand_only.xlsx')

 
# $$$$$$$$$$$$AHMED GHORAIB$$$$$$$$$$ رسمتين Plotly جنب بعض =========
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('توزيع الفئات (من المُصنَّف)', 'عدد المنتجات لكل فئة'),
    specs=[[{"type": "pie"}, {"type": "bar"}]]
)

colors = ['#00CC00','#66CCFF','#FFD700','#FFB347',
          '#FF9900','#FF4444','#AAAAAA','#4444FF']

# Pie $$$$$$$AHMED GHORAIB$$$$$$$$$من المُصنَّف
if classified > 0:
    fig.add_trace(
        go.Pie(
            labels=labels,
            values=vals,
            marker=dict(colors=colors),
            text=[f"{l}<br>{v} من {classified} ({round(v/classified*100,2)}%)"
                  for l, v in zip(labels, vals)],
            textinfo='text'
        ),
        row=1, col=1
    )
else:
    fig.add_trace(
        go.Pie(labels=['لا يوجد بيانات مصنَّفة'], values=[1]),
        row=1, col=1
    )

# $$$$$$$$$$AHMED GHORAIB$$$$$$Bar بعدد المنتجات
fig.add_trace(
    go.Bar(
        x=labels,
        y=vals,
        marker_color=colors,
        text=vals,
        textposition='outside'
    ),
    row=1, col=2
)

fig.update_layout(
    height=600,
    title="تحليل ترتيب سعر: من أرخص لحد خامس + أغلى + وسط + عندي بس",
    showlegend=False
)

fig.show()
# this is anew coding in python for helped me Ai Tools for describe him ######
#Coding To Run BY $$$$$$$$$$$$$AHMED GHORAIB$$$$$$$$$ IN 20/1/2026
```
